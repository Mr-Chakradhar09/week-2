## HISTOGRAM
It helps in visualizing the distribution of each specified feature individually through histograms.
It's useful for understanding the spread, central tendency, and shape of the data for each feature separately.

```python
# Distribution of 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall' features
features = ['N','P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']
fig, ax=plt.subplots(3,3,figsize=(15,15))
for i ,subplot in zip(features,ax.flatten()):
     sns.histplot(crop[i],ax=subplot, kde=True)
     subplot.set_title(i)
#plt.title(f'Distribution of {feature}')
plt.tight_layout()
plt.show()
```

```python
#ploting scatter plot
import seaborn as sns
fig, ax = plt.subplots(3, 3, figsize=(15,15))
for i, subplot in zip(features,ax.flatten()):
    sns.scatterplot(x=i,y='label',data=crop,ax=subplot)
plt.tight_layout()
plt.show()
```

# BOXPLOT
* It helps visualize the distribution and statistical summary of each specified feature using boxplots.
* Boxplots provide information about the median, quartiles, and potential outliers in the data.

```python
# Boxplot for 'N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall' features
features = ['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']
for feature in features:
    sns.boxplot(x=crop[feature])
    plt.title(f'Boxplot of {feature}')
plt.tight_layout()
plt.show()
```

```python
fig, ax=plt.subplots(3,3,figsize=(15,15))
for i, subplot in zip(features,ax.flatten()):
    sns.boxplot(crop[i],ax=subplot)
plt.tight_layout()
plt.show()
```

1. Whiskers extend to the minimum and maximum values within 1.5 times the IQR from the first and third quartiles, respectively.
2. Points outside the whiskers are considered outliers and plotted individually.
3. It allows you to visually identify the central tendency, spread, and potential outliers in each feature's distribution.

### Bivariate Analysis
Bivariate analysis is crucial in understanding the relationship between two different variables.

```python
# Setting plot aesthetics
sns.set_theme(style="darkgrid")
plt.rcParams['figure.figsize'] = [10, 8]


# Box Plots for a variable against crop types
sns.boxplot(x='label', y='N', data=crop)
plt.xticks(rotation=90)
plt.title('Nitrogen Content across Different Crops')
plt.show()
```
